```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

### CV text - Luís Vale

```python
text = """
Mestrado Green Data Science ISA Instituto Superior Agronomia Lisboa
Licenciatura Engenharia Florestal Recursos Naturais ISA Instituto Superior Agronomia Lisboa
Camara Municipal Lisboa fiscalização Equipa Arvoredo monitorizando regulamentação ambiental
melhoria Técnicas instalação Arvoredo podas formação projetos base natural Arboricultura Urbana
sensibilização ambiental climática Estrutura Verde Arvoredo urbano
Avaliação extração nemátodes INIAV Instituto Nacional Investigação Agrária Veterinária
Manutenção culturas puras protocolos esterilização controlo qualidade autoclavagem amostras
madeira normas laboratoriais Registo procedimentos resultados investigação científica
Programação Análise Dados Python R SQL Excel MariaDB MySQL
Jupyter Notebooks DBeaver Git GitHub PowerPoint
Microscopia técnicas esterilização protocolos segurança laboratorial
Português Inglês Florestal Ambiental Urbano Arvoredo Natural
Data Science Python R análise dados visualização ambiente florestal recursos naturais
investigação laboratório microscopia qualidade científica biodiversidade ecologia
sustentabilidade conservação florestas árvores verde urbano
"""
```

### Define stopwords

```python
stopwords = set(STOPWORDS)

stopwords.update([
    "Luis", "Vale", "Luís", "Henrique", "Simões", "Correia", "Moreira",
    "ISA", "Lisboa", "INIAV", "DMAEVCE", "NMP", "SONAE",
    "de", "da", "do", "dos", "das", "em", "na", "no", "nos", "nas",
    "e", "o", "a", "os", "as", "com", "por", "para", "que", "se",
    "ao", "um", "uma", "ou", "Colaboração", "Ferramentas",
    "Estágio", "Instituto", "Nacional", "Municipal", "Microsoft"
])
```

### Generate and display the word cloud

```python
wordcloud = WordCloud(
    stopwords=stopwords,
    max_font_size=80,
    max_words=100,
    background_color="white",
    width=800,
    height=500,
    colormap="Greens",   # Green tones - fitting for Forestry and Environment!
    collocations=False
).generate(text)

plt.figure(figsize=(12, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Word Cloud - Luís Vale CV", fontsize=16, pad=15)
plt.tight_layout()
plt.savefig("wordcloud_LuisVale.png", dpi=150, bbox_inches='tight')
plt.show()
print("Saved as wordcloud_LuisVale.png")
```

### Questions

**Q1: How many variables is the word cloud based on?**

The word cloud is based on **2 variables**: the **word itself** and its **frequency** (encoded as font size).

**Q2: What is the type of each variable?**

- **Word** → Qualitative / Nominal (categorical, no natural order)
- **Frequency** → Quantitative / Ratio (numerical, encoded visually through font size)
